```python
import pandas as pd

# Initialize variables to keep track of where the metadata ends
metadata_end_line = 0

with open(file_path, 'r') as file:
    for line_number, line in enumerate(file):
        if not line.startswith('!'):
            metadata_end_line = line_number
            break

df = pd.read_csv(file_path, delimiter='\t', skiprows=range(0, metadata_end_line), index_col=0)

df.head()
```

```python
# drop rows that appear to be additional metadata (rows starting with '!')
df_cleaned = df.drop(df.index[df.index.str.startswith('!')])

# clean data
df_cleaned = df_cleaned.apply(pd.to_numeric, errors='coerce')

df_cleaned.head()
```
